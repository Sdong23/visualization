```python
# Load library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv("D:/UT_DataScience/assignments/visualization/02_activities/assignments/private_school_contact_information_eng.csv")
```

```python
# Sort data
region_counts = df['Region'].value_counts().sort_index()
total_schools = region_counts.sum()
percentages = (region_counts / total_schools * 100).round(2)
```

```python
# Raw Plot
plt.figure(figsize=(10,4))
df['Region'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.show()
```

```python
# Modify bar plot
ax1 = plt.gca()
bars = ax1.bar(
    region_counts.index, 
    region_counts.values,
    color='#4CAF50', 
    alpha=0.7,
    label='School Count'
)
```

```python
# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
# Fig 1
ax1.bar(region_counts.index, region_counts.values, color='#4CAF50', alpha=0.7)
ax1.set_title("Schools of region")
ax1.set_ylabel("Number of schools")
# add labels
for bar in bars:
    height = bar.get_height()
    ax1.text(
        x=bar.get_x() + bar.get_width()/2, 
        y=height + 0.2,                    
        s=f'{int(height)}',                  
        ha='center',                         
        va='bottom',                         
        fontsize=10,
        color='black'
    )
# change x axis
labels = [label.replace(' ', '\n') for label in region_counts.index]
ax1.set_xticklabels(labels, rotation=45, ha='center')
# Fig 2
ax2.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=90)
ax2.set_title("Percent of total schools")
# Plot show
plt.tight_layout()
plt.show()
```
